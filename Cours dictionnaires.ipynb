Nous utilisons les…………………………... {} pour définir le début et la fin du dictionnaire (alors que nous utilisons des crochets [ ] pour les tableaux et des parenthèses ( ) pour les tuples).

Dans le dictionnaire ci-dessus, …

<b><u>Important</u></b>...


* Les valeurs peuvent être des chaînes de caractères, des booléens....

* Le dictionnaire est une représentation en Python d'une fonction (au sens mathématique) qui a une clé associe une valeur.
                        f : {"nom", "prenom", "date de naissance"} →  {chaînes de caractères}
                           
                                                             "nom"   |→  "Durand"
                                                             
                                                          "prenom"   |→  "Christophe"
                                                          
                                                "date de naissance"  |→  "29/01/2001"
